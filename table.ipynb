```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [3.3, 0, 2.6]
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, color='green')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatterplot')

ax.set_xticks([])
#ax.set_yticks([])
```

```python
plt.show()
```

```python
y1 = [2.3] * len(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1, label="Dados", color='purple', linestyle='-')
```
